```python
import pandas as pd
col = ['id','imdb_id','popularity','budget_adj','revenue_adj','homepage',
       'keywords','overview','production_companies','vote_count','vote_average']
#row = [0,1,2,3,4,5,6,7,8]
# deleting the columns.
tmdbdata = pd.read_csv('dataset.csv')
print(len(col))
#print(len(row))
tmdbdata = tmdbdata.drop(col,1)
#tmdbdata = tmdbdata.drop(row,0)
tmdbdata
```

```python
#removing the duplicates.

tmdbdata.drop_duplicates(keep = 'first',inplace=True)
# printing the dataset
tmdbdata
```

```python
col = ['budget','revenue','runtime']
col1 = ['budget','revenue']
# to replace 0 with NAN
tmdbdata[col] = tmdbdata[col].replace(0,np.NAN)

# to drop NAN value from budget and revenue column
tmdbdata.dropna(subset=col1,inplace = True)

#printing the dataset

tmdbdata
```

```python
tmdbdata.dtypes
```

### Which movie had the most and least profit

```python
# insert the column name profit
tmdbdata.insert(2,'profit',tmdbdata['revenue']-tmdbdata['budget'])
```

```python
def analyses(column):
    high = tmdbdata[column].idxmax()
    highrow = pd.DataFrame(tmdbdata.loc[high])
    print(high)
    low = tmdbdata[column].idxmin()
    lowrow = pd.DataFrame(tmdbdata.loc[low])
    print(low)
    analysis1 = pd.concat([highrow,lowrow],axis =1)
    return analysis1
```

```python
analyses('profit')
```

# Year of release vs Profitability

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
tmdbdata = pd.read_csv('dataset.csv')
tmdbdata.insert(2,'profit',tmdbdata['revenue']-tmdbdata['budget'])
profitsyear = tmdbdata.groupby('release_year')['profit'].sum()
print(profitsyear.head())
print(profitsyear.idxmax())
```

```python
profitdata = tmdbdata[tmdbdata['profit'] >= 60000000]

profitdata.index = range(len(profitdata))#giving new indexing to all rows

#profitdata.index = profitdata.index + 1 #starting from 1
profitdata.head()
```

# Successful Genre

```python
def data(col):
    data = profitdata[col].str.cat(sep = "|") #concatenating using pipeline ("|")
#     print(data)
    data = pd.Series(data.split('|')) #spliting in a pd series
#     print(data)
    count = data.value_counts(ascending = False) 
    return count
count = data('genres')
count
```

```python
def data(col):
    data = profitdata[col].str.cat(sep = "|") #concatenating using pipeline ("|")
#     print(data)
    data = pd.Series(data.split('|')) #spliting in a pd series
#     print(data)
    count = data.value_counts(ascending = False) 
    return count
count = data('genres')
count
df = pd.DataFrame(count)
ax = df.plot(kind='pie',subplots = True, figsize=(20,20),
             title='Pie Plot on Genres', fontsize=15)
```

# Most Frequent Cast

```python
count = data('cast')
count.head()
```

```python
dta = count.head(10)
dta.sort_values(ascending = True, inplace = True)

lt = dta.plot.barh(color = '#0000EF', fontsize = 13) #horrizontal bar graph
lt.set(title= 'Most Frequent Cast')
lt.set_xlabel('No of movies in the datasey', color = 'black', fontsize = '13')
lt.set_ylabel('cast',color='black',fontsize = '13')
lt.figure.set_size_inches(12,6)
plt.show()
```
